```python
# Higher levels of accuracy when training data shas been denormalized
import numpy
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten

from keras.layers import Dropout
from keras.utils import np_utils

# load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
img_width = X_train.shape[1]
img_height = X_train.shape[2]

X_train = X_train.astype('float32')

# Normalizing the data dramatically improves the predictions
X_train /= 255.

# one hot encode outputs
y_train = np_utils.to_categorical(y_train)
num_classes = y_train.shape[1]

# create model
model2=Sequential()
model2.add(Flatten(input_shape=(img_width,img_height)))

# number of neurons in the network
# model.add(Dense(30, activation='relu'))

model2.add(Dense(num_classes, activation='softmax'))
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model2.fit(X_train, y_train)
```

```python
# Using actual testing data in normalized form
X_test = X_test.astype('float32')
X_test /= 255.
y_test = np_utils.to_categorical(y_test)
model2.fit(X_test, y_test)
```
